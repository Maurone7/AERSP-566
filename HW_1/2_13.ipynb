```python
import numpy as np
import matplotlib.pyplot as plt

func_1 = lambda x: 0.5 if x>-1 and x<1 else 0

x = np.linspace(-1.5,1.5,1000)
func_1_matrix = [func_1(i) for i in x]

def func_2(x):
    if x>-1 and x<0:
        return x+1
    elif x>0 and x<1:
        return -x+1
    else:
        return 0
    
func_2_matrix = [func_2(i) for i in x]
plt.plot(x, func_1_matrix)
plt.show()

plt.plot(x, func_2_matrix)
plt.show()
```

```python
Z_matrix = [(func_1_matrix[i] + func_2_matrix[i])/2 for i in range(len(x))]

plt.plot(x, Z_matrix)
```
