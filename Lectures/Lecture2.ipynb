## Uniform

$f_x(x)=\frac{1}{b-a}$ if $x \epsilon [a, b]$

```python
import numpy as np
import matplotlib.pyplot as plt

function = lambda a, b, x: 1 / (b - a) if a <= x <= b else 0

a, b = 0, 1
# Plot the uniform distribution
plt.plot(np.linspace(a, b, 100), [function(a, b, x) for x in np.linspace(a, b, 100)])
plt.ylabel('Frequency')
plt.title('Uniform Distribution')
plt.ylim(0, 2 / (b - a)), plt.xlim(a-0.5, b+0.5)
plt.yticks([0, 1 / (b - a)], [0, r'$\frac{1}{(b - a)}$'])
plt.xticks([a, b], ['a', 'b'])
plt.show()
```

## Gaussian 
#### (or normal distribution)

$f_x(x)=\frac{1}{\sigma \sqrt{2\pi}}e^{\frac{-(x-\bar{x})^2}{2\sigma^2}}$

$x \sim N(\bar{x}, \sigma^2)$

The "N" stand for normal distribution

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parameters of the normal distribution
mean = 0
std = 1

# Generate data points from the normal distribution
x = np.linspace(-3, 3, 100)
y = (1 / (std * np.sqrt(2 * np.pi))) * np.exp(-((x - mean) ** 2) / (2 * std ** 2)) 

plt.plot(x, y)
plt.ylabel('Probability Density')
plt.title('Normal Distribution')
plt.xticks([-3*std, 0, 3*std], [r'$\bar{x}-3\sigma$', r'$\bar{x}$', r'$\bar{x}+3\sigma$'])
plt.yticks([])
plt.show()
```

#### Probability distribution

$F_x(x)=\frac{1}{\sigma \sqrt{2\pi}}\int_{-\infty}^{\infty}e^{\frac{-(z-\bar{x})^2}{2\sigma^2}}dz$

### Symmetric PDF
If $f_x(x)$ is even ($f_x(x)=f_x(-x)$) $\rightarrow$ ith moment $=E(x^i)=\int_{-\infty}^{\infty}x^if_x(x)dx=\int_{-\infty}^{0}x^if_x(x)dx+\int_{0}^{\infty}x^if_x(x)dx$

If $f_x(x)$ is odd ($x^i=-(-x)^i$) $\rightarrow \int_{-\infty}^{0}x^if(x)dx=\int_{0}^{\infty}(-x)^if(-x)dx=-\int_{0}^{\infty}x^if(x)dx$ \
$i^{th}$ moment=0 for odd values of i
